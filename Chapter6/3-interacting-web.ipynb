# 6.3 Interacting with Web APIs

```python
import numpy as np
import pandas as pd
```

```python
import requests
```

```python
url = 'https://api.github.com/repos/pandas-dev/pandas/issues'
```

```python
resp = requests.get(url)
```

```python
resp
```

```python
data = resp.json() 
# return list then a dictionary containing JSON parsed into native's python object
```

```python
type(data)
```

```python
data[0]['title']
```

```python
issues = pd.DataFrame(data, columns=['number', 'title', 'labels', 'state'])
```
