## Barplots

```python
fig, axes = plt.subplots(2, 1)
data = pd.Series(np.random.rand(16), index=list('abcdefghijklmnop'))
data.plot.bar(ax=axes[0], color='k', alpha=0.7)
data.plot.barh(ax=axes[1], color='k', alpha=0.7);
```

```python
tips = pd.read_csv('tips.csv')
```

```python
tips.head()
```

```python
tips['tip_pct'] = tips['tip'] / (tips['total_bill'] - tips['tip'])
```

```python
tips.head()
```

```python
sns.barplot(x='tip_pct', y='day', data=tips, orient='h')
# the black line is the 95% confidence interval
```

```python
sns.barplot(x='tip_pct', y='day', data=tips, hue='time', orient='h')
```

## Histograms and Density Plots

```python
tips['tip_pct'].plot.hist(bins=50)
```

```python
tips['tip_pct'].plot.density()
# kernel density estimate
```

```python
# using seaborn
comp1 = np.random.normal(0, 1, size=200)
comp2 = np.random.normal(10, 2, size=200)
values = pd.Series(np.concatenate([comp1, comp2]))
sns.distplot(values, bins=100, color='k')
```

## Scatter or Point Plots

```python
macro = pd.read_csv('macrodata.csv')
```

```python
macro.head()
```

```python
data = macro[['cpi', 'm1', 'tbilrate', 'unemp']]
```

```python
trans_data = np.log(data).diff().dropna()
```

```python
trans_data[:5]
```

```python
sns.regplot('m1', 'unemp', data=trans_data)
plt.title('Changes in log %s versus log %s' % ('m1', 'unemp'))
plt.grid();
```

```python
sns.pairplot(trans_data, diag_kind='kde', plot_kws={'alpha' : 0.2})
```

## Facet Grids and Categorical Data

Facet Grids for visualize data with many categorical variables.

```python
sns.factorplot(x='day', y='tip_pct', hue='time', col='smoker', kind='bar',
              data=tips[tips.tip_pct < 1])
```

```python
sns.factorplot(x='day', y='tip_pct', data=tips[tips.tip_pct < 1],
              row='time', col='smoker', kind='bar')
```

```python
sns.catplot(x='tip_pct', y='day', kind='box', data=tips[tips.tip_pct < 0.5])
```
